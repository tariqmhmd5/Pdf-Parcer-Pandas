```python
#1302 +1 pages
import numpy as np
import pandas as pd
import pdfplumber
import tabula
import re
from pandas import ExcelWriter
from pandas import ExcelFile
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.DataFrame(np.nan,index=range(2500),columns=['College Name','University','Branch','GOPENS','GOPENH'])
```

```python
pdf = pdfplumber.open('college_cutoffs.pdf') #to read the file text
file = "college_cutoffs.pdf"
count = 0          # pointer to point indexes in dataframe (to append new row)

start_time = time.time()
for page in range(1303):
    table = tabula.read_pdf(file,pages=page+1)
    page = pdf.pages[page]
    page_text = page.extract_text()
    
    br_df = pd.DataFrame(re.findall(r'\d{9} [-] ([^\n\r]+)', page_text),columns=['Branch'])
    clg_df = pd.DataFrame(re.findall(r'\b\d{4} [-] ([^\n\r]+)', page_text),columns=['College'])
    uni_df = pd.DataFrame(re.findall(r' : ([^\n\r]+)', page_text),columns=['University'])
    
    
    c=0
    indexes=[]
    for i in range(len(table)):
        if 'GOPENS' in list(table[i].columns):
            indexes.append(i)
            c+=1
        elif 'GOPENH' in list(table[i].columns):
            c+=1
            indexes.append(i)
                                  
    if c!=0:
        try:
            i=0 #for branch
            for index in indexes[:br_df.shape[0]]:
                if 'GOPENS' in list(table[index].columns):
                    df['GOPENS'][count] = table[index]['GOPENS'][0]
                elif 'GOPENH' in list(table[index].columns):    
                    df['GOPENH'][count] = table[index]['GOPENH'][0]
                df['University'][count] = uni_df['University'][0]
                df['College Name'][count] = clg_df['College'][0]
                df['Branch'][count] = br_df['Branch'][i]
                i+=1
                count+=1

        except Exception as e:
            print(e)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
writer = ExcelWriter('Cleaned Data.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```

# Data Cleaning

```python
df = pd.read_excel('Cleaned Data.xlsx')
```

```python
data = df.head(1662)
```

```python
data.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
data.fillna('-',inplace=True)
```

```python
for i in range(len(data['GOPENH'])):
    if len(data['GOPENH'][i])>5:
        data['GOPENH'][i] = data['GOPENH'][i].split('\r')[0]
```

```python
for i in range(len(data['GOPENS'])):
    if len(data['GOPENS'][i])>5:
        data['GOPENS'][i] = data['GOPENS'][i].split('\r')[0]
```

```python
data['GOPENH'] = data['GOPENH'].replace('-','0')
data['GOPENS'] = data['GOPENS'].replace('-','0')
data['GOPENH'] = pd.to_numeric(data['GOPENH'])
data['GOPENS'] = pd.to_numeric(data['GOPENS'])
```
